```python
#FEATURE EXTRACTION:tfidf
```

```python
import re
```

```python
def wordopt(text):
  text = text.lower()
  #remove urls
  text = re.sub(r'https?://\S+|www\.\S+','',text)
  # remove html tags
  text = re.sub(r'<.*?>','',text)

  #removing punctuation
  text = re.sub(r'[^\w\s]', '',text)

  # removing digits
  text = re.sub(r'\d','',text)

  #removing newline characters
  text = re.sub(r'\n', ' ', text)

  return text

```

```python
news['text'] = news['text'].apply(wordopt)
```

```python
news['text']
```

```python
news.head()
```

```python
x = news['text']
y = news['label']
```

```python
print(x)
```

```python
print(y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3,random_state = 0)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorization = TfidfVectorizer()
```

```python
xv_train = vectorization.fit_transform(x_train)
```

```python
xv_test = vectorization.transform(x_test)
```

```python
print(xv_train)
```

```python
print(xv_test)
```

```python
#CLASSIFICATION MODEL
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression(random_state=0)
```

```python
LR.fit(xv_train,y_train)
```

```python
print("Shape of xv_train:", xv_train.shape)
print("Shape of xv_test:", xv_test.shape)
```

```python
lr_pred = LR.predict(xv_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
cn = confusion_matrix(y_test,lr_pred)
print(cn)
print(accuracy_score(y_test, lr_pred))
print(classification_report(y_test,lr_pred))
```

```python
#DECISSION TREE CLASSIFIER
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
dtc.fit(xv_train, y_train)
```

```python
y_pred = dtc.predict(xv_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix, accuracy_score
cn = confusion_matrix(y_test,y_pred)
print(cn)
print(accuracy_score(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

```python
#RANDOM FOREST CLASSIFIER
from sklearn.ensemble import RandomForestClassifier
rclf = RandomForestClassifier()
rclf.fit(xv_train,y_train)
print("The Accuracy of the Random Forest Classifier Model is {}".format(rclf.score(xv_test,y_test)))
print(classification_report(y_test,rclf.predict(xv_test)))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gclf = GradientBoostingClassifier()
gclf.fit(xv_train,y_train)
print("The Accuracy of the Decision Tree Classifier Model is {}".format(gclf.score(xv_test,y_test)))
print(classification_report(y_test,gclf.predict(xv_test)))
```

```python
def output_lable(n):
    if n == 0:
        return "Fake News"
    elif n == 1:
        return "Not A Fake News"

def manual_testing(news):
    testing_news = {"text":[news]}
    new_def_test = pd.DataFrame(testing_news)
    new_def_test["text"] = new_def_test["text"].apply(wordopt)
    new_x_test = new_def_test["text"]
    new_xv_test = vectorization.transform(new_x_test)
    pred_LR = LR.predict(new_xv_test)
    pred_DT = dtc.predict(new_xv_test)
    pred_GBC = gclf.predict(new_xv_test)
    pred_RFC = rclf.predict(new_xv_test)

    return print("\n\nLR Prediction: {} \nDT Prediction: {} \nGBC Prediction: {} \nRFC Prediction: {}".format(output_lable(pred_LR[0]),                                                                                                       output_lable(pred_DT[0]),
                                                                                                              output_lable(pred_GBC[0]),
                                                                                                              output_lable(pred_RFC[0])))
```

```python
news = str(input())
manual_testing(news)
```
